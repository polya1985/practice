```python
n=int(input(''))
T=[]
for k in range(n):
    T.append(input().split())
for s in range(n):
    p=int(T[s][0])
    n=int(T[s][1])
    print()
    for j in range(n):
        t=0
        
        for i in range(p):
            t=t+1
            print(f"{t}"*t)
        for k in range(p-1):
            t=t-1
            print(f"{t}"*t)
        print()
```
